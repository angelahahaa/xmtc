```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import scipy.sparse
import collections
```

```python
def make_ft(x):
    t = x['sequence']
    t = t[t>0]
    t = ' '.join(t.astype(str))
    l1 = ' '.join(['__label__' + n for n in x['categories']])
    l2 = '__label__' + x['hierarchy'][0]
    return [' '.join([l1,t]),' '.join([l2,t])]
```

# from raw data

```python
df = pd.read_pickle('data/sic_hierarchy.pkl')
```

```python
df.head()
```

```python
labs_c = df.categories.apply(lambda x: ' '.join(['__label__' + n for n in x]))
labs_h = df.hierarchy.apply(lambda x:'__label__' + x[0])
seqs = df.sequence.apply(lambda x:' '.join(x[x>0].astype(str)))
df['FT_c'] = labs_c + ' ' + seqs
df['FT_h'] = labs_h + ' ' + seqs
```

```python
df.head()
```

```python
for mode in ['train','test']:
    np.savetxt('data/FT/sic_c.{}.txt'.format(mode),df[df['train/test']==mode].FT_c.values,fmt='%s')
    np.savetxt('data/FT/sic_h.{}.txt'.format(mode),df[df['train/test']==mode].FT_h.values,fmt='%s')    
```
